```python
def imputing_missing_values(dataset):
    dataset.category_name.fillna('Unknown', inplace=True)
    dataset.brand_name.fillna('Unknown', inplace=True)
    dataset.item_description.fillna('Unknown', inplace=True)
    return dataset

def preprocess_text(text_string):
    words_preprocessing_filters = [strip_punctuation, strip_multiple_whitespaces, remove_stopwords, stem_text]
    return preprocess_string(text_string, words_preprocessing_filters)

def dataframe_merge_text_features(dataframe, list_of_preprocessed_text_features, merged_text_feature_name):
    new_dataframe = dataframe.copy()
    text_features_accessors_list = [''.join(['row["', text_feature, '"]']) for text_feature in list_of_preprocessed_text_features]
    string_of_lambda_fn = '+'.join(text_features_accessors_list)
    new_dataframe[merged_text_feature_name] = new_dataframe.apply(lambda row: eval(string_of_lambda_fn), axis=1)
    new_dataframe[merged_text_feature_name] = new_dataframe.apply(lambda row: ' '.join(row[merged_text_feature_name]), axis=1)
    return new_dataframe
```

Fill missing values with unknown tags and perform text preprocessing on textual features

```python
data_X = data.drop('price', axis=1)
data_y = data.loc[:, 'price']
data_X = imputing_missing_values(data_X)
data_X['name'] = data_X['name'].apply(preprocess_text)
data_X['item_description'] = data_X['item_description'].apply(preprocess_text)
```

Do sampling of 10000 records (training set), 2000 records (validation set) and 2000 records (test set)

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(data_X, data_y, train_size=12000, test_size=2000, random_state=15)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, train_size=10000, test_size=2000, random_state=15)
```

## Exploratory Data Analysis

Since there are too many variations for the 'category-name' column, we decided to split the categories up, focusing on 1 main category and 2 sub-categories instead.

```python
def split_str(text, index):
    try:
        cats_subcats = text.split('/')
        return cats_subcats[index]
    except:
        # Only consider 1 main cat and 2 sub-cats
        return None
```

```python
X_train['main_category'] = X_train['category_name'].apply(split_str, args=(0,))
X_train['sub_category_1'] = X_train['category_name'].apply(split_str, args=(1,))
X_train['sub_category_2'] = X_train['category_name'].apply(split_str, args=(2,))
X_train.head()
```

## Topic Modeling with Latent Dirichlet Allocation
This is a type of statistical modeling for discovering the abstract "topics" that occur in a collection of document.</br>
Compared to Normal distribution, Dirichlet distribution is also a probability distribution but over a space of real numbers, imagine a list of numbers that add up to 1.

```python
def get_word_list(text):
    return text.split()

X_train['name_description'] = X_train['name_description'].apply(get_word_list)
X_train.head(5)
```

```python
# Creating a dictionary of our corpus
from gensim import corpora
dictionary = corpora.Dictionary(X_train['name_description'])

# filter_extremes() to filter out tokens (less than 20 documents, more than 0.5 fraction of total corpus, first 100000 tokens)
dictionary.filter_extremes(no_below=20, no_above=0.5, keep_n=100000)

# Converting list of documents into Document-Term Matrix
# Each list contains of words in doc, in the form of (index in dictionary, no. of word occurence)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in X_train['name_description']]
```

```python
# Creating object for LDA model
import gensim
LDA = gensim.models.ldamodel.LdaModel

# Running LDA model
ldamodel = LDA(doc_term_matrix, num_topics=10, id2word=dictionary, passes=50)
```
